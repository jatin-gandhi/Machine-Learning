```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test ,y_train ,y_test =train_test_split(df.drop(columns=['user','churn'],axis=1),df['churn'],test_size = 0.2)
```

```python
y_train.value_counts()
```

```python
pos_index = y_train[y_train==1].index
neg_index = y_train[y_train==0].index
```

```python
if len(pos_index) > len(neg_index):
    higher = pos_index
    lower =neg_index
else:
    higher=neg_index
    lower=pos_index
```

```python
import random
```

```python
random.seed(0)
```

```python
higher = np.random.choice(higher,size = len(lower))
lower = np.asarray(lower)
```

```python
new_index = np.concatenate((lower,higher))
```

```python
X_train = X_train.loc[new_index,:]
```

```python
y_train = y_train[new_index]
```

```python
from sklearn.preprocessing import  StandardScaler
```

```python
sc = StandardScaler()
```

```python
scale_x = sc.fit_transform(X_train)
```

```python
scale_xtest = sc.fit_transform(X_test)
```

```python
scaled_train = pd.DataFrame(scale_x,columns=X_train.columns)
```

```python
scaled_test = pd.DataFrame(scale_xtest,columns=X_test.columns)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(scaled_train,y_train)
```

```python
pred = classifier.predict(scaled_test)
```

```python
from sklearn.metrics import classification_report ,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
coeff =pd.concat([pd.DataFrame(scaled_train.columns,columns=['Features']),pd.DataFrame(np.transpose(classifier.coef_),columns=['Coefficient'])],axis=1)
```
